**Track/Untrack files**

✅`Tracking`:In Git, "tracking files" means that Git is monitoring these files for changes. When a file is tracked, any changes to it are recorded by Git and can be included in commits to save the changes to the project's history.


👉*Key points`
▶Tracked Files : Files that Git is monitoring for changes.

▶Untracked Files : Files that Git is not monitoring yet.

▶Staging Area : Where changes are placed before they are committed.

▶Commit : Saving changes to the repository's history.

👉**Commands**

▶git add: Starts tracking a new file or stages changes to a tracked file.`git add filename` or `git add .`(to track all files at a time) or `git add --all` or `git add -A`

▶git status: Shows which files are tracked, untracked, and their state (modified, staged).`git status`

▶git commit: Records the staged changes to the repository's history.`git commit -m "Commit message"` ,here `-m` Stands for "message". It specifies that the following string should be used as the commit message.

▶git rm --cached: Stops tracking a file but keeps it in the working directory.`git rm --cached filename`

**Examples**

▶Adding and Tracking a New File:

`touch newfile.txt`# Create a new file

`git add newfile.txt` # Start tracking the file

`git commit -m "Add newfile.txt"`  # Commit the file

▶Use a .gitignore file to tell Git which files to ignore (not track).

`# Example .gitignore file`

`*.log`

`*.tmp`

`node_modules/`

✅`untrack files`:In Git, "untracking files" means telling Git to stop monitoring changes to specific files. This is often done when you no longer want certain files to be included in version control, either because they are unnecessary for the repository or they contain sensitive information that should not be tracked.

👉To stop tracking a file but keep it in the working directory, you use the `git rm --cached command`

👉Example:`git rm --cached config.json`

▶This command removes config.json from the staging area and tells Git to stop tracking it. The file will still remain in your working directory but will no longer be tracked by Git.


👉Example scenario:

▶Start Tracking a File:
`echo "Sensitive data" > secret.txt`

`git add secret.txt`

`git commit -m "Add secret.txt"`

▶Realize the Mistake and Untrack the File:

`git rm --cached secret.txt`

`echo "secret.txt" >> .gitignore`

`git commit -m "Stop tracking secret.txt and add to .gitignore"`

**Delete and remove the files**

✅In Git, if you want to delete and remove files from the repository, you can use the `git rm command`. This command removes the files from your working directory and stages the deletions for the next commit.

👉How to Delete and Remove Files in Git:

▶1.Delete a Single File:`git rm filename`

▶Example:git `rm unwanted_file.txt`

`git commit -m "Remove unwanted_file.txt"`

▶This command deletes `unwanted_file.txt` from both the working directory and the repository, and stages the change for the next commit.

▶Delete Multiple Files:

`git rm old_file1.txt old_file2.txt`

`git commit -m "Remove old and unused files"`

▶Remove Files But Keep Them Locally:If you want to stop tracking a file but keep it in your local working directory, use the `--cached` option:

▶`git rm --cached filename`

▶Example:`git rm --cached local_config.json`

`echo "local_config.json" >> .gitignore`

`git commit -m "Stop tracking local_config.json and add to .gitignore"`

▶This command removes `local_config.json` from the repository but leaves the file in your local directory. Adding it to .gitignore ensures Git does not track it again.

▶To restore deleted files:`git restore "file name"`

▶If we want to change file name in git use this `git mv "old_name" "new_name"`

▶To know the history of commits ,simply type `git log` we get all the history commits in detail.If we need to see all commits in single_line then use `git log --oneline`

▶To view the chnages in commits simply use `git lo -p` it shows all the changes we did in commits

▶If we need to reset any changes of file use`git reset filename`

**Branches**

✅Branches in Git are a way to diverge from the main line of development and continue to work without affecting that main line. They are used to develop features, fix bugs, or experiment in an isolated environment. When changes in a branch are ready, they can be merged back into the main branch (often called `main` or `master`)

👉We create another branch that basically copy of main branch with same entities like main branch.Use this before fixing bugs.

▶To create new branch `git branch branch_name`

▶To see branches on our git `git branch` it gives us all brnaches we have.

▶To switch the branch use the command `git switch branch_name`

▶To merge back changes done in one branch into main branch `git merge -m "merge" branch name` (branch what we need to merge with main)

▶After making all this changes and merges in branches ,to delete the branch use the command `git branch -d branch_name`

▶If we have feature bug in work then we create another branch and make all changes in that branch then merge it back into main branch
